OriginalLowFilingPrice: The lower end of the initially proposed price range for the IPO shares. The distribution is similar to the high filing price, indicating the range that companies expect to sell their shares.

OriginalMiddleFilingPriceRange: The midpoint of the original filing price range. It follows the same pattern as the high and low, skewed towards the lower end.

SharesFiledinthisMkt: The number of shares that were filed to be sold in this market. The histogram indicates that most companies file to sell a relatively modest number of shares.

SharesFiledsumofallMkts: The total number of shares filed for sale across all markets. Again, most IPOs involve fewer shares, with some exceptions.

AmtFiledinthisMktml: This could represent the amount (in millions) filed for in this particular market, similar to the proceeds but relating to the filing.

TotalDollarAmountFiled: The total dollar amount filed for the IPO. Most companies have lower total filed amounts, with a few higher values.

TotalSharesOfferedInThisMarke: This probably indicates the total number of shares actually offered in this market, which can be different from the number filed due to changes between filing and offering.

PrimarySharesOfferedinthisMkt: The number of primary (new) shares offered in this market. The distribution is skewed, suggesting that most offerings are small.

SecondaryShrsOfferedinthisMkt: The number of secondary shares (previously issued shares sold by existing shareholders) offered in this market.

SharesOfferedsumofallMkts: The total number of shares offered across all markets.

# Creating the taget variable - Underpricing

```python
#creating the underpricing variable
df['underpricing'] = (df['YesterdayStockPrice'] - df['OfferPrice']) / df['OfferPrice']

# Display the first few rows to verify the new variable
print(df[['YesterdayStockPrice', 'OfferPrice', 'underpricing']].head())
```

```python
# Count the number of underpricing values below zero
count_below_zero = (df['underpricing'] < 0).sum()

# Count the number of underpricing values above zero
count_above_zero = (df['underpricing'] > 0).sum()

# Count the total number of non-null underpricing values
total_count = df['underpricing'].notnull().sum()

# Calculate the percentages
percentage_below_zero = (count_below_zero / total_count) * 100
percentage_above_zero = (count_above_zero / total_count) * 100

# Display the percentages
print(f"Percentage of underpricing values below zero: {percentage_below_zero}%")
print(f"Percentage of underpricing values above zero: {percentage_above_zero}%")
```

```python
# Scatter plot for 'PrincipalAmountmil'
plt.figure(figsize=(10, 6))
plt.scatter(main.index, main['PrincipalAmountmil'], alpha=0.5)
plt.title('Scatter Plot of Principal Amount (mil)')
plt.xlabel('Index')
plt.ylabel('Principal Amount (mil)')
plt.show()
```

```python
mean = main['PrincipalAmountmil'].mean()
std =  main['PrincipalAmountmil'].std()

cut_off = std * 3
lower, upper = mean - cut_off, mean + cut_off
trimmed_main = main[(main['PrincipalAmountmil'] < upper) &
        (main['PrincipalAmountmil'] > lower)]
```

```python
# Scatter plot for 'PrincipalAmountmil'
plt.figure(figsize=(10, 6))
plt.scatter(trimmed_main.index, trimmed_main['PrincipalAmountmil'], alpha=0.5)
plt.title('Scatter Plot of Principal Amount (mil)')
plt.xlabel('Index')
plt.ylabel('Principal Amount (mil)')
plt.show()
```

A commonly used alternative approach is to remove data that sits further than three standard deviations from the mean. You can implement this by first calculating the mean and standard deviation of the relevant column to find upper and lower bounds, and applying these bounds as a mask to the DataFrame. This method ensures that only data that is genuinely different from the rest is removed, and will remove fewer points if the data is close together.

# LightGBM

```python
import lightgbm as lgb
from sklearn import metrics
import numpy as np


# Initialize lists to store results
rocs_lgbm = []
lgbm_models = []
l_rate = []
n_leaves = []
max_depths = []
subsamples = []

# Define hyperparameter ranges
learning_rates = [0.001, 0.01, 0.03, 0.05, 0.1, 0.15, 0.3]
num_leaves = [15, 20, 31, 63, 70]
max_depth_range = [3, 5, 7, 10]
subsample_range = [0.6, 0.8, 1.0] # Fraction of samples to use

# Fixed parameters
params = {
    'boosting_type': 'gbdt', # type of alogrithm to run
    'objective': 'binary', # binary classification problem
    'is_unbalance': 'true', # to match for the imbalance dataset
    'metric': 'auc', # metric to evaluate is ROC AUC
    'random_state': 0, # to replicate results
}

for rate in learning_rates: # controls the step size at each iteration
    params['learning_rate'] = rate
    for l in num_leaves: # to control complexity of the model
        params['num_leaves'] = l
        for max_depth in max_depth_range: # limits the maximum tree depth
            params['max_depth'] = max_depth
            for subsample in subsample_range: #subsample ratio of the training instance
                params['subsample'] = subsample

                # Initialize and train model
                gbm = lgb.LGBMClassifier(n_estimators=500, **params) #number of boosting iterations 500
                gbm.fit(train_scaled, y_train, eval_set=[(dev_scaled, y_dev)],
                        callbacks=[lgb.callback.early_stopping(10, verbose=5), #regularisation to avoid overfitting
                                   lgb.callback.log_evaluation(50)]) # logs the evaluation results of the model

                # Make predictions
                y_dev_pred_proba_lgbm = gbm.predict_proba(dev_scaled, num_iteration=gbm.best_iteration_)

                # Evaluation
                fpr, tpr, thresholds = metrics.roc_curve(y_dev, y_dev_pred_proba_lgbm[:, 1])
                roc_auc_lgbm = metrics.auc(fpr, tpr)

                # Store results
                rocs_lgbm.append(roc_auc_lgbm)
                lgbm_models.append(gbm)
                l_rate.append(rate)
                n_leaves.append(l)
                max_depths.append(max_depth)
                subsamples.append(subsample)

# Identify the best model
best_index_lgbm = np.argmax(rocs_lgbm)
best_auc_lgbm = rocs_lgbm[best_index_lgbm]
best_model_lgbm = lgbm_models[best_index_lgbm]

print(f"Best AUC: {best_auc_lgbm}")
print(f"Best Learning Rate: {l_rate[best_index_lgbm]}")
print(f"Best Num Leaves: {n_leaves[best_index_lgbm]}")
print(f"Best Max Depth: {max_depths[best_index_lgbm]}")
print(f"Best Subsample: {subsamples[best_index_lgbm]}")
```

```python
# Define model

def createModel():
  tf.keras.utils.set_random_seed(42)
  model = Sequential([
    Dense(128, activation='relu', kernel_initializer=HeNormal(), input_shape=(train_scaled.shape[1],)),
    Dense(64, activation='relu', kernel_initializer=HeNormal()),
    Dense(32, activation='relu', kernel_initializer=HeNormal()),
    Dense(1)  # A single output
  ])
  return model

# create model
model = createModel()
```
